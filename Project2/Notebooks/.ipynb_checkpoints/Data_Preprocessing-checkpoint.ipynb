```python
A=[]
B=[]
for i in target_freq.keys():
    
    if i[0]=='A':
        A.append(i)
    else:
        B.append(i)
        
print(len(A),A)
print()
print(len(B),B)
```

```python
B_target=[]
int_target=[]

for i in train['target']:
    
  
    if i[0]=='A':
        B_target.append(1)
        int_target.append(int(i[1:]))
    else:
        B_target.append(0)
        int_target.append(-int(i[1:]))
len(B_target)
```

```python
int_target
```

```python
train["B_target"]=B_target
train["int_target"]=int_target
```
